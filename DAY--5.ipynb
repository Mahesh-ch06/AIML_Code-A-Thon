# Import necessary libraries



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
# Load dataset
data = pd.read_csv('/content/survey lung cancer.csv')
```

```python
data
```

```python
data.head()
```

```python
data.info()
```

# Remove any leading or trailing whitespace from column names

```python
data.columns = data.columns.str.strip()
```

```python
data.columns
```

# Display the first few rows and column names to confirm cleanup

```python
print("First few rows of the dataset:")
data.head()
```

```python
print("\nColumn names:", data.columns)
```

# Check for any missing values

```python
print(data.isnull().sum())
```

# Data Preprocessing - Encoding categorical variables if any exist

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
data['GENDER'] = label_encoder.fit_transform(data['GENDER'])
data['LUNG_CANCER'] = label_encoder.fit_transform(data['LUNG_CANCER'])
```

# Separate features and target variable

```python
X = data.drop('LUNG_CANCER', axis=1)
y = data['LUNG_CANCER']
```

# Split data into training and testing sets (test_size=0.5, random_state=42)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

# Initialize the Random Forest Classifier (n_estimators=100, random_state=42)

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

# Train the model

```python
rf_classifier.fit(X_train, y_train)
```

# Make predictions

```python
y_pred = rf_classifier.predict(X_test)
```

# Evaluate the model

```python
print("\nModel Accuracy:", accuracy_score(y_test, y_pred))
print("\nModel Accuracy:", accuracy_score(y_test, y_pred)*100) # Converting into percentage
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

# Feature importance Graph

```python
feature_importances = rf_classifier.feature_importances_
features = X.columns
```

# Plotting feature importances

```python
# Plotting feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances, y=features)
plt.title("Feature Importance in Random Forest Model")
plt.xlabel("Importance")
plt.ylabel("Features")
plt.show()
```
