# Import necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the data

```python
# Load dataset
data = pd.read_csv('/content/column_2C_weka.csv')
```

# Step 1: Data Preprocessing

```python
label_encoder = LabelEncoder()
data['class'] = label_encoder.fit_transform(data['class'])
X = data.drop(columns=['class'])
y = data['class']
```


# Step 2: Standardize the Features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Split the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# Step 3: Linear Regression


**Train Linear Regression model**

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

## Predict and calculate MSE

```python
y_pred_linear = linear_model.predict(X_test)
mse_linear = mean_squared_error(y_test, y_pred_linear)
print(f"Linear Regression MSE: {mse_linear}")
```

## Plot predicted vs actual

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_linear, color='blue', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression: Predicted vs Actual')
plt.show()
```

# Step 4: Logistic Regression

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
```



# Predict and display Confusion Matrix

```python
y_pred_logistic = logistic_model.predict(X_test)
cm_logistic = confusion_matrix(y_test, y_pred_logistic)
ConfusionMatrixDisplay(cm_logistic).plot()
plt.title('Confusion Matrix for Logistic Regression')
plt.show()
```

# Step 5: Decision Tree Classifier

```python
# Train Decision Tree Classifier
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)
```

# Feature Importance Plot

```python
plt.figure(figsize=(10, 6))
feature_importance = tree_model.feature_importances_
sns.barplot(x=feature_importance, y=X.columns)
plt.title('Feature Importance for Decision Tree')
plt.show()
```

# Step 6: Random Forest Classifier

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

# Confusion Matrix for Random Forest

```python
y_pred_rf = rf_model.predict(X_test)
cm_rf = confusion_matrix(y_test, y_pred_rf)
ConfusionMatrixDisplay(cm_rf).plot()
plt.title('Confusion Matrix for Random Forest')
plt.show()
```

# Step 7: K-Nearest Neighbors

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

# Plotting the Distance Plot

```python
distances, indices = knn_model.kneighbors(X_test[:1], n_neighbors=5)
plt.plot(range(1, 6), distances.flatten(), marker='o')
plt.xlabel('Neighbor Rank')
plt.ylabel('Distance')
plt.title('K-Nearest Neighbors Distance Plot')
plt.show()
```

# Step 8: Support Vector Machine (SVM)

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
# Confusion Matrix for SVM
y_pred_svm = svm_model.predict(X_test)
cm_svm = confusion_matrix(y_test, y_pred_svm)
ConfusionMatrixDisplay(cm_svm).plot()
plt.title('Confusion Matrix for SVM')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Calculate and print the Mean Squared Error for Linear Regression
mse_linear = mean_squared_error(y_test, y_pred_linear)
print(f"Linear Regression: Mean Squared Error (MSE) of {mse_linear * 100:.2f}")

# Calculate and print the accuracy for Logistic Regression
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print(f"Logistic Regression: Accuracy of {accuracy_logistic * 100:.2f}%")

# Calculate and print the accuracy for Decision Tree
accuracy_tree = accuracy_score(y_test, tree_model.predict(X_test))
print(f"Decision Tree Classifier (DT): Accuracy of {accuracy_tree * 100:.2f}%")

# Calculate and print the accuracy for Random Forest
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Classifier (RF): Accuracy of {accuracy_rf * 100:.2f}%")

# Calculate and print the accuracy for K-Nearest Neighbors
accuracy_knn = accuracy_score(y_test, knn_model.predict(X_test))
print(f"K-Nearest Neighbors (KNN): Accuracy of {accuracy_knn * 100:.2f}%")

# Calculate and print the accuracy for Support Vector Machine
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"Support Vector Machine (SVM): Accuracy of {accuracy_svm * 100:.2f}%")
```
