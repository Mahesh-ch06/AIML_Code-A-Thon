# import libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the Dataset

```python
# Load dataset
data = pd.read_csv('/content/car_evaluation_codethon.csv')
```

```python
data
```

```python
data.head()
```

# Encode categorical data

```python
X = data.drop('decision', axis=1) #features
y = data['decision'] #target
```

# One-hot encode categorical features in X

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

# Define features (X) and target (y)

```python
X = data.drop('decision', axis=1)
y = data['decision']
```

# Split the dataset into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.7, random_state=42)
```

# Initialize and train the Decision Tree Classifier

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

# Predict on the test data

```python
y_pred = clf.predict(X_test)
```

# Calculate accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

#  Generate classification report

```python
report = classification_report(y_test, y_pred)
```

```python
print(report)
```

# Plot confusion matrix as a heatmap

```python
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix - Decision Tree Classifier')
plt.show()
```

```python
# Displaying results
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print(f"Accuracy: {accuracy*100:.2f}") # Converting into percentage
print("\nClassification Report:\n", report)
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
submission = pd.DataFrame({
    'Accuracy': [accuracy]
})
submission = submission[['Accuracy']]
submission.to_csv('DAY-4.csv', index=False)
print("Submission file 'DAY-4.csv' created successfully.")
```
